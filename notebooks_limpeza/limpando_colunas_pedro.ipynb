```python
import pandas as pd


df17 = pd.read_parquet('DENGBR17.parquet')
df18 = pd.read_parquet('DENGBR18.parquet')
df19 = pd.read_parquet('DENGBR19.parquet')

df = pd.concat([df17, df18, df19], ignore_index=True)

colunas_pedro = [
    'ID_RG_RESI', 'COMPLICA', 'DT_NOTIFIC', 'GENGIVO', 'ID_PAIS', 'DT_GRAV',
    'ALRM_SANG', 'DT_CHIK_S2', 'RES_CHIKS2', 'SANGRAM', 'DOENCA_TRA', 'ID_AGRAVO',
    'ALRM_VOM', 'DT_ENCERRA', 'DT_SORO', 'GRAV_EXTRE', 'FEBRE', 'NU_ANO',
    'DT_CHIK_S1', 'PLAQ_MENOR', 'CS_SEXO', 'ALRM_HIPOT', 'TPAUTOCTO', 'HOSPITALIZ',
    'ID_UNIDADE', 'ID_OCUPA_N', 'EVOLUCAO', 'GRAV_MELEN', 'NAUSEA', 'DT_VIRAL',
    'NU_IDADE_N', 'ALRM_LETAR', 'DIABETES', 'CS_FLXRET', 'MUNICIPIO', 'SG_UF',
    'CRITERIO', 'ID_REGIONA'
]

colunas_target = [
    'ID_RG_RESI', 'DT_NOTIFIC', 'ID_AGRAVO', 'DT_ENCERRA',
    'FEBRE', 'NU_ANO', 'CS_SEXO', 'TPAUTOCTO', 'HOSPITALIZ',
    'ID_UNIDADE', 'ID_OCUPA_N', 'NAUSEA',
    'ID_REGIONA', 'SG_UF', 'NU_IDADE_N'
]

df_filtrado = df[colunas_target].copy()

def parse_idade(valor):
    try:
        s = str(int(valor)).zfill(4)
        unidade = int(s[0])
        numero  = int(s[1:])
        if unidade == 4:
            return numero
        elif unidade == 3:
            return numero / 12
        elif unidade == 2:
            return numero / 365
        elif unidade == 1:
            return numero / 8760
        else:
            return None
    except:
        return None

df_filtrado['IDADE_ANOS'] = df_filtrado['NU_IDADE_N'].apply(parse_idade)
df_filtrado = df_filtrado.drop(columns=['NU_IDADE_N'])


df_filtrado['DT_NOTIFIC'] = pd.to_datetime(df_filtrado['DT_NOTIFIC'], format='%Y-%m-%d')
df_filtrado['MES_NOTIFIC'] = df_filtrado['DT_NOTIFIC'].dt.month
df_filtrado['DIA_NOTIFIC'] = df_filtrado['DT_NOTIFIC'].dt.day
df_filtrado = df_filtrado.drop(columns=['DT_NOTIFIC'])

df_filtrado['DT_ENCERRA'] = pd.to_datetime(df_filtrado['DT_ENCERRA'], format='%Y-%m-%d')
df_filtrado['MES_ENCERRA'] = df_filtrado['DT_ENCERRA'].dt.month
df_filtrado['DIA_ENCERRA'] = df_filtrado['DT_ENCERRA'].dt.day
df_filtrado = df_filtrado.drop(columns=['DT_ENCERRA'])


df_filtrado = df_filtrado.dropna(subset=['FEBRE', 'ID_OCUPA_N', 'TPAUTOCTO'])

print(list(df_filtrado))
display(df_filtrado)
#print(df_filtrado.shape)
#print(df_filtrado.isnull().sum())



```
